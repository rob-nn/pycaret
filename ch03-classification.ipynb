```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
from pycaret.datasets import get_data
from pycaret.classification import *
mpl.rcParams['figure.dpi'] = 300
```

```python
data = get_data('iris')
```

```python
data.info()
```

```python
data['species'].value_counts().plot(kind='pie')
plt.ylabel('')
plt.show()
```

```python
fig, axes = plt.subplots(2,2, figsize=(12,10))

for ax, col in zip(axes.flatten(), data.columns):
    sns.boxplot(data = data, x='species', y = col, ax=ax)
    ax.set_xlabel('')
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(data.corr().round(decimals=2), annot=True)
plt.show()
```

```python
sns.pairplot(data, hue='species')
plt.show()
```

```python
classf = setup(data = data, target='species', train_size=0.8, normalize=True, session_id=3934)
```

```python
compare_models(sort='Accuracy')
```

```python
model = create_model('lda')
```

```python
model_cat = create_model('catboost', verbose=False)
```

```python
params = {
    'iterations': np.arange(100, 1000, 100),
    'max_depth': np.arange(1, 10),
    'learning_rate': np.arange(0.01, 1, 0.01),
    'random_strength': np.arange(0.01, 1, 0.01),
    'l2_leaf_reg': np.arange(1, 100),
    'border_count': np.arange(1 , 256)
}
```

```python
tuned_model = tune_model(model_cat, optimize='Accuracy', fold=10, tuner_verbose=False, search_library = 'scikit-optimize', custom_grid=params, n_iter=50)
```
